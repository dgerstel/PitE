```python
fl_data = pd.read_csv('../../Tacview-CSV-Samples/A-4E (Viper) [Red].csv', sep=',', encoding='latin1', index_col='Time', parse_dates=['Time'])
fl_data[1500:]
```

```python
from mpl_toolkits.basemap import Basemap
```

```python
lon = fl_data['Longitude'].tolist()
lat = fl_data['Latitude'].tolist()

EPS = np.std(lat)
llc_lat = min(lat) - EPS
llc_lon = min(lon) - EPS
urc_lat = max(lat) + EPS
urc_lon = max(lon) + EPS
```

```python
themap = Basemap(projection='merc',
              llcrnrlon = llc_lon,              # lower-left corner longitude
              llcrnrlat = llc_lat,               # lower-left corner latitude
              urcrnrlon = urc_lon,               # upper-right corner longitude
              urcrnrlat = urc_lat,               # upper-right corner latitude
              resolution = 'l',
              area_thresh = 100000.0,
              lon_0 = 0.5*(llc_lon+urc_lon),
              lat_0 = 0.5*(llc_lat+urc_lat)
              #satellite_height = 30000.0
              )
```

```python
themap.drawcoastlines()
themap.drawcountries()
themap.fillcontinents(color = 'gainsboro')
themap.drawmapboundary(fill_color='steelblue')
```

```python

x, y = themap(lon, lat)
themap.plot(x, y, 
            'o',                    # marker shape
            color='Indigo',         # marker colour
            markersize=4            # marker size
            )

plt.show()
```
