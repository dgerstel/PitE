### Project organisation

The projected is contained within the main class _AppMgr_, with the following classes embedded:
- _DataGenerator_
- _Fitter_
- _StatAnalyser_
- _Plotter_


#### _DataGenerator_
Its role is to generate data according to model provided and to simulate gaussian noise N(0, $\sigma$) in either of two forms:
- proportional -- the bigger the y-value the greater the noise. The motivation behind it is an inherent noise of a physical process, e.g. Poisson distribution based fluctuations of a number of nuclei that disintegrate in a unit time.
The outcome is, then, (1 + N(0, $\sigma$) multiplied by pure model.

- white noise -- it's purely random and pertains to uncertainty of measurement as opposed to physical phenomenon intrinsic fluctuations.
The outcome is, then, pure model plus N(0, $\sigma$)


The y values are computed for equidistant x arguments, so no x error is assumed.


#### _Fitter_
It's an object that exploits *curve_fit* function from *scipy.optimize* module. It is used twice per analysis: at first to fit model to data and then to fit gaussian to residual distributions obtained after fitting the data.


#### _StatAnalyser_
It embeds two methods: *residuals* and *chi2ndf*.
The latter calculates chi squared between the data and fit and yields that value divided by number of degrees of freedom (no. of data points minus one). The former, however, complements that test as far as symmetry of errors is concerned -- here we obtain values that are sensitive to positive/negative error sign.


#### _Plotter_
This class enables plotting data and fit superimposed. It is, similarly to _Fitter_ , used twice: for the sample data and for the residuals. It plots x and y data either as points or as histogram bars. In the latter case, it is assumed the histogram bins and their contents are provided as x and y data. This strategy was chosen instead of *pyplot.hist* method, which is a tool for both computing and ploting a histogram, in order to separate these two actions. In this way, _Plotter_ is free from any computing and performs just data visualisation as one would expect.



#### Project execution
Project execution is invoked by _AppMgr_'s *_\_init\_\_* method in the subsequent order:
- a sample model is assumed -- $sin(x)$ was used by picking specific parameters of more general *self.sin_linear*, which is $a \cdot sin(bx) + c$, so this could be easily altered. The model *self.sin_linear* is a method within the class just to store everything within the main class _AppMgr_. Similar strategy has been applied to other models: *gaussian*, *polyn_n*.

- lists of fit models and their parameters is created. They contain two elements: same fit model as sample model (the difference is, however, that sample has noise) and 3-rd degree polynomial.

- then, just for convenience x-axis is quantised and limited

- then data are generated iterating over various noise levels

- for each noise level data are generated accordingly and iteration over fit models is implemented

- single analysis (for specific noise and fit model) comprises: fitting model to the data, calculating residuals and chi squared and plotting the results on two plots: data with their fit and histogrammed residuals with their fit. The details are described in the above sections.



```python

class AppMgr(object):


    
    def __init__(self, noise=0.3, xRange=(0, np.pi*2), sampleSize=1e2):
        
        # assume model of data and its parameters
        model =  self.sin_linear #self.sin_with_noise #
        pars = [1, 1, 0] # [1, 0, 1, 1, 0] #

        # various fit models and initial guess of their params
        fit_models = [model, self.polyn_n]#self.sin_with_noise
        fit_pars_arr = [pars, (1, 1, 1, 1)]#[1, 0, 1, 1, 0]
        
        # prepare x-axis
        axis = np.linspace(xRange[0], xRange[1], sampleSize)        

        
        # various noise levels
        for noise in np.linspace(.1, 0.6, 6):
            
            # generate data (with gaussian noise) using the model
            dg = self.DataGenerator((model,pars), axis, noise, noiseType="white")
            data = dg.generate()
            x, y = data[:, 0], data[:, 1]

            for i, fitMdlWithPars in enumerate(zip(fit_models, fit_pars_arr)):
                
                fit_model, fit_pars = fitMdlWithPars[0], fitMdlWithPars[1]
                
                # fit the model to the data; intercept statistics data
                ft = self.Fitter(fit_model, x, y, fit_pars)
                popt, pcov = ft.fit()


                # get fit_y points
                fit_y = fit_model(x, *popt)

                # calc residuals and chi squared / ndf
                sta = self.StatAnalyser(y, fit_y)
                resid = sta.residuals()
                chi2_per_ndf = sta.chi2ndf()
                
                # compute histogram of the residuals
                hist, bin_edges = np.histogram(resid, bins=np.linspace(-6, 6, 100),
                                               normed=True)
                bin_cntr = bin_edges[:-1] + 0.5 * (bin_edges[1] - bin_edges[0])
                
                # fit gaussian to the residuals
                res_ft = self.Fitter(self.gauss, bin_cntr, hist, (1,0,1))
                res_ft_parm, pcov = res_ft.fit()
                res_mu, res_sig = res_ft_parm[1], res_ft_parm[2]

                fit_res_y = self.gauss(bin_cntr, *res_ft_parm)

                
                # plot data with fit and residuals and chi squared / ndf

                label='$\sigma$ = %1.2f' % noise
                model_name = fit_model.__name__

                dat_tit = "Data and fit for noise param. " + label + " and model " + model_name
                res_tit = "Histogram of residuals for noise param. " + label + " and model " + model_name
                
                chi2_info = '$\chi^{2} / ndf =$ %1.2f' % (chi2_per_ndf)
                res_fit_info = '$\mu = %1.2f$ \n $\sigma = %1.2f$' % (res_mu, res_sig)

                # plot data with fit
                data_pl = self.Plotter(data=(x,y), fit=(x, fit_y), hist=False, title=dat_tit,
                                       info=chi2_info)
                # plot residuals with gaussian fit
                res_pl = self.Plotter(data=(bin_edges, hist), fit=(bin_cntr, fit_res_y), hist=True,
                                      title=res_tit, info=res_fit_info)
                
        
        
    def sin_linear(self, x, a, b, c):
        """
        
        model function
        
        """
        return a * np.sin(b * x) + c

    
    def gauss(self, x, *p):
        N, mu, sig = p
        return N * 1./(sig * np.sqrt(2*np.pi)) * np.exp(-1./2 * ((x-mu)/sig)**2 )
    
    def polyn_n(self,x, *p):
        """
        polynomial of nth-order
        """
        
        pol = np.poly1d(p)
        return pol(x)
    
    
    class DataGenerator(object):
        """

        generates data according to the params:
        - trueModel -- f(x), e.g. np.sin 
        - noiseLevel -- sigma of normal distribution N(0, sigma)
        - xRange -- (x_min, x_max) tuple for x in [x_min ; x_max] range
        - sampleSize -- number of points returned
        - norm -- coefficient the resulting values are multiplied by

        The values returned are generated for equidistant arguments within xRange.
        Maybe x-values should be chosen randomly?
        Also, maybe add x-error?
        What about y-errors? Do we know them? Ignore them?

        """

        def __init__(self, modelWithPars, xaxis, noiseLevel, noiseType="white", norm=1.0):
            
            self.model, self.pars = modelWithPars[0], modelWithPars[1]
            self.axis = xaxis
            self.noise = noiseLevel
            self.norm = norm
            self.smpSize = len(self.axis)
            self.noiseType = noiseType


        def generate(self):
            """
            generates sample according to self.model and stores y values in the object, so they
            can be accessed later on

            returns 2-dim numpy array of x,y data
            """

            self.y = self.norm * (self.model(self.axis, *self.pars))
            noise = np.random.normal(0, self.noise, int(self.smpSize))                
                                  
            if self.noiseType == "white":
                self.y = self.y + noise
                
            elif self.noiseType == "proportional":
                self.y = self.y * (1 + noise)
                                 
            return np.array( zip(self.axis, self.y) )


    class Fitter(object):
        
        def __init__(self, model, xdata, ydata, p0):
            
            self.model = model
            self.xdata = xdata
            self.ydata = ydata
            self.p0 = p0
               
        def fit(self):
            from scipy.optimize import curve_fit
            
            return curve_fit(self.model, self.xdata, self.ydata, self.p0 )
        
        

    class StatAnalyser(object):
        
        """
        calculates (y_i - y_fit)^2 """
        
        def __init__(self, data_y, model_y):
            
            self.data = data_y
            self.model = model_y
            
        
        def residuals(self):
            """
            
            calculates residuals normalised to mean value of data and model:
            res_i = (data_i - model_i) / model_i
            
            """
            
            return (self.data - self.model) / self.model 
        
        def chi2ndf(self):
            return np.sqrt( ((self.data - self.model) ** 2 / abs(self.model)).sum() / len(self.data) )
        

    class Plotter(object):
        """

        Plots data assumed to be 2-dim numpy array

        """

        def __init__(self, data, fit, hist=False, title="", info=""):
            """
            
            data is 2-dim numpy array of x and generated y
            fit_model is a tuple of fit function and its params
            
            """
            self.x, self.y = data[0], data[1]
            self.fit_x, self.fit_y = fit[0], fit[1]
            self.hist = hist
            self.title = title
            self.info = info
                      
            self.plot()

                      
        def plot(self):
            
            fig, ax = plt.subplots(1, 1, figsize=(12, 10), sharex = False, sharey=False)
            
            # plot data
            if self.hist:
                ax.bar(self.x[:-1], self.y, width=(self.x[1] - self.x[0]), alpha=0.5)
            else:
                ax.plot(self.x, self.y, 'bo', label='data')  
            
            # plot fit
            ax.plot(self.fit_x, self.fit_y, 'r-', label='fit') 
            
                      
            # cosmetics
#             if self.title is not None:
            ax.set_title(self.title)
                      
            ax.text(.6 * max(self.x), .6 * max(self.y), self.info, ha='center',
                     size=18, alpha=0.5, color='red') 

            ax.legend()
            
            
    
```

### Analysing the results
We can see that increasing the noise spreads out residuals distribution and increases $\chi^2$ per number of degrees of freedom, as expected. At $\sigma=0.50$ the $\chi^2 / ndf$ is greater than 1 in both fit models used (sine and 3-rd order polynomial). It means, the models stop being reliable as the noise increases and perhaps a better model would be a convolution of signal and noise.
